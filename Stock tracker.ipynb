```python
stock_prices = {"AAPL":180,"TSLA":250,"INFY":1500,"TCS":3200,"GOOG":140}
print("Welcome to Stock Portfolio Tracker")
print("Available_stocks:",list(stock_prices.keys()))
stock = input("Enter the stock symbol you want to buy:").upper()
if stock not in stock_prices:
    print("Sorry, this stock is not available")
else:
    quantity = int(input("Enter the quantity:"))
    price = stock_prices[stock]
    total = price * quantity

    print(f"\nStock Selected: {stock}")
    print(f"Price per share: {price}")
    print(f"Quantity:{quantity}")
    print(f"Total Investment: \u20B9{total}")
    
 #Saving a file
    with open("investment_report.txt","w",encoding="utf-8")as file:
        file.write(f"Stock Portfolio Summary\n")
        file.write(f"Stock:{stock}\n")
        file.write(f"Price per share:{price}\n")
        file.write(f"Quantity:{quantity}\n")
        file.write(f"Total Investment: \u20B9{total}\n")
        
    print("\nYour report has been saved as 'investment_report.txt' ")
```
